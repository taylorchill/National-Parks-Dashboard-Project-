```python
import urllib.request, json

base_url = "https://developer.nps.gov/api/v1/parks?"
api_key = nps_key
```

```python
indices = [0,50,100,150,200,250,300,350,400,450]
```

```python
url= f"{base_url}&api_key={api_key}"

response = requests.get(url).json()
```

```python
data = response['data']

name = data[0]['fullName']

images = data[0]['images']

print (images)
```

```python
images_df = pd.DataFrame(images)
images_df
```

```python
images_df['park'] = name
images_df
```

```python
imagesdata = []


for index in indices: 
    url = (f"{base_url}&start={index}&api_key={api_key}")
    response = requests.get(url).json()

    data = response['data']

    for i in range(len(data)):
        
        images = data[i]['images']
        
        park = data[i]['fullName']
        
        images_df = pd.DataFrame(images)
        
        images_df['park'] = park
        
        imagesdata.append(images_df)
```

```python
all_images_df = pd.concat(imagesdata)
all_images_df.head(20)
```

```python
all_images_df.count()
```

```python
all_images_df = all_images_df.reset_index()
```

```python
clean_images_df = all_images_df[["caption", "url", "park"]]

clean_images_df.head(30)
```

```python

engine = sqlalchemy.create_engine('sqlite:///../nps.sqlite', echo=False)

all_images_df.to_sql('all_images_df', con=engine, if_exists = "replace")
```
