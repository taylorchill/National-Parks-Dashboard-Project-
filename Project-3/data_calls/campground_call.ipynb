```python
import requests
import pandas as pd
import ast
import sqlalchemy
from sqlalchemy import Column, String, Float, Integer
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import func
from config import nps_key
```

```python
import urllib.request, json

base_url = "https://developer.nps.gov/api/v1/campgrounds?"
api_key = nps_key
```

```python
indices = [0,50,100,150,200,250,300,350,400,450,500,550,600]

campgrounddata = []
amenitiesdata = []
imagesdata = []
feesdata = []
campsitesdata = []
```

```python
for index in indices: 
    url = (f"{base_url}&start={index}&api_key={api_key}")
    response = requests.get(url).json()

    data = response['data']

    for i in range(len(data)):
             
        campground = data[i]['name']
        
        amenities = data[i]['accessibility']
        
        images = data[i]['images']
        
        fees = data[i]['fees']
        
        campsites = data[i]['campsites']

        campsites_df = pd.DataFrame(campsites, index= [0])

        campsites_df['name'] = campground

        campsitesdata.append(campsites_df)
        
        amenities_df = pd.DataFrame(amenities)
        
        amenities_df['name'] = campground
        
        amenitiesdata.append(amenities_df)
        
        images_df = pd.DataFrame(images)
        
        images_df['campground'] = campground
        
        imagesdata.append(images_df)
        
        fees_df = pd.DataFrame(fees)
        
        fees_df['name'] = campground
               
        feesdata.append(fees_df)
        
        entry = data[i]

        campgrounddata.append(entry)
```

```python
campground_df = pd.DataFrame(campgrounddata)
campground_df.head()
```

```python
campground_df["image"] = campground_df["images"].apply(lambda x: x[0]["url"] if len(x) else 0)
campground_df.head()
```

```python
all_amenities_df = pd.concat(amenitiesdata)
all_amenities_df.head(20)
```

```python
all_amenities_df.columns
```

```python
all_fees_df = pd.concat(feesdata)
all_fees_df.head(20)
```

```python
all_images_df = pd.concat(imagesdata)
all_images_df.head(20)
```

```python
all_campsites_df = pd.concat(campsitesdata)
all_campsites_df.head(20)
```

```python
camp_df = campground_df.loc[:, ['name', 'parkCode', 'description', 'latitude', 'longitude', 'reservationUrl', 'weatherOverview',
        
       'numberOfSitesReservable',
       'numberOfSitesFirstComeFirstServe', 'image']]

camp_df.head()
```

```python
camp_w_amenities_df = pd.merge(camp_df, all_amenities_df, how = "left", on = "name" )

camp_w_amenities_df.head()
```

```python
all_camp_df = pd.merge(camp_w_amenities_df, all_campsites_df, how = "left", on = "name" )

all_camp_df.head()
```
