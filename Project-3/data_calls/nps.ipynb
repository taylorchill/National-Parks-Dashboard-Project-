```python
import urllib.request, json

base_url = "https://developer.nps.gov/api/v1/parks?"
api_key = nps_key
```

```python
indices = [0,50,100,150,200,250,300,350,400,450]

parksdata = []
```

```python
for index in indices: 
    url = (f"{base_url}&start={index}&api_key={api_key}")
    response = requests.get(url).json()

    data = response['data']

    for i in range(len(data)):

        entry = data[i]

        parksdata.append(entry)     
```

```python
nps_df = pd.DataFrame(parksdata)

nps_df.head()
```

```python
nps_df.columns
```

```python
clean_df = nps_df.loc[:, ["url", "fullName", "parkCode", "description", "latitude", "longitude", "activities", 
                           "states", "entranceFees", "directionsInfo", "images", "weatherInfo"]]
clean_df.head(10)
```

```python
clean_df["activity"] = clean_df["activities"].apply(lambda x:','.join( [i["name"] for i in x]))
clean_df.head()
```

```python
clean_df["image"] = clean_df["images"].apply(lambda x: x[0]["url"] if len(x) else 0)
clean_df.head()
```

```python
clean_df["cost"] = clean_df["entranceFees"].apply(lambda x: x[0]["cost"] if len(x) else 0)
clean_df.head()
```

```python
clean_df = clean_df.loc[:, ["url", "fullName", "parkCode", "description", "latitude", "longitude", "activity", 
                           "states", "cost", "directionsInfo", "image", "weatherInfo"]]
clean_df.head()
```

```python

engine = sqlalchemy.create_engine('sqlite:///../nps.sqlite', echo=False)


clean_df.to_sql('clean_df', con=engine, if_exists = "replace")
```
